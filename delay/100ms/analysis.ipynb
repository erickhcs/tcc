```python
import pandas as pd
import os

basePath = '.'
fileName = 'user1-session1_QoE.csv'
```

```python
data_merged = pd.read_csv(basePath + '/' + '1' + '/' + fileName)
```

```python
for i in range(2, 2476):
  path = basePath + '/' + str(i) + '/' + fileName
  if os.path.exists(path):
    data = pd.read_csv(path)
    data_merged = pd.concat([data_merged,data], ignore_index=True)


data_merged.tail(5)
```
