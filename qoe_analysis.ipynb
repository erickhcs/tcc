```python
import pandas as pd
import os

folderClient_prefix = 'client_'

def run_analysis(basePath):
  final_data = pd.DataFrame()
  for i in range(1, 5):
    for j in range(1, 15):
        for k in range(1, 8):
          try:

            path = basePath +  str(i) + '/client_' + str(k) + '/' + str(j) + '/user1-session1_QoE.csv'
            if not os.path.exists(basePath):
              continue
            data = pd.read_csv(path)
            final_data = pd.concat([final_data, data], ignore_index=True)

          except Exception as a:
            pass
  
  return final_data

```
