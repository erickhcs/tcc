```python
import pandas as pd

basePath = './0-0-0/client_'
data_merged = pd.DataFrame()

for i in range(1, 5):
  for j in range(1, 834):
    data = pd.read_csv(basePath + str(i) + '/' + str(j) + '/user1-session1_QoE.csv')
    data_merged = pd.concat([data_merged, data], ignore_index=True)
```
