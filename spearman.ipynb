```python
import pandas as pd
import os

folderClient_prefix = 'client_'

basePath000 = './0-0-0/' + folderClient_prefix
basePath003 = './0-0-3/' + folderClient_prefix
finalFileBasePath = 'user1-session1_QoE.csv'

def run_analysis(basePath):
  final_data = pd.DataFrame()
  for i in range(1, 5):
    for j in range(1, 834):
      data = pd.read_csv(basePath + str(i) + '/' + str(j) + '/user1-session1_QoE.csv')
      final_data = pd.concat([final_data, data], ignore_index=True)

  return final_data
  
```

```python
data000 = run_analysis(basePath000)
data003 = run_analysis(basePath003)

data_merged = pd.concat([data000, data003], ignore_index=True)
```
