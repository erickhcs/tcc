```python
import pandas as pd

basePathDelay1 = './delay/default'
basePathDelay2 = './delay/100ms'
basePathDelay3 = './delay/300ms'
basePathDelay4 = './delay/500ms'
basePathDelay5 = './delay/100ms'

basePathJitter1 = './jitter/20ms'
basePathJitter2 = './jitter/50ms'
basePathJitter3 = './jitter/100ms'
basePathJitter4 = './jitter/200ms'

basePathLoss1 = './loss/0.1%'
basePathLoss2 = './loss/1%'
basePathLoss3 = './loss/3%'
basePathLoss4 = './loss/5%'

finalFileBasePath = 'user1-session1_final.csv'

pathDelay1 = basePathDelay1 + '/1/' + finalFileBasePath
pathDelay2 = basePathDelay2 + '/2/' + finalFileBasePath
pathDelay3 = basePathDelay3 + '/3/' + finalFileBasePath
pathDelay4 = basePathDelay4 + '/4/' + finalFileBasePath
pathDelay5 = basePathDelay5 + '/5/' + finalFileBasePath

pathJitter1 = basePathJitter1 + '/1/' + finalFileBasePath
pathJitter2 = basePathJitter2 + '/2/' + finalFileBasePath
pathJitter3 = basePathJitter3 + '/3/' + finalFileBasePath
pathJitter4 = basePathJitter4 + '/4/' + finalFileBasePath

pathLoss1 = basePathLoss1 + '/1/' + finalFileBasePath
pathLoss2 = basePathLoss2 + '/2/' + finalFileBasePath
pathLoss3 = basePathLoss3 + '/3/' + finalFileBasePath
pathLoss4 = basePathLoss4 + '/4/' + finalFileBasePath
```

```python
dataDelay1 = pd.read_csv(pathDelay1)
dataDelay2 = pd.read_csv(pathDelay2)
dataDelay3 = pd.read_csv(pathDelay3)
dataDelay4 = pd.read_csv(pathDelay4)
dataDelay5 = pd.read_csv(pathDelay5)

dataJitter1 = pd.read_csv(pathJitter1)
dataJitter2 = pd.read_csv(pathJitter2)
dataJitter3 = pd.read_csv(pathJitter3)
dataJitter4 = pd.read_csv(pathJitter4)

dataLoss1 = pd.read_csv(pathLoss1)
dataLoss2 = pd.read_csv(pathLoss2)
dataLoss3 = pd.read_csv(pathLoss3)
dataLoss4 = pd.read_csv(pathLoss4)
```

```python
data_merged = pd.concat([dataDelay1,dataDelay2,dataDelay3,dataDelay4,dataDelay5,dataJitter1,dataJitter2,dataJitter3,dataJitter4,dataLoss1,dataLoss2,dataLoss3,dataLoss4], ignore_index=True)


data_merged.describe()
```

```python
data_merged = data_merged.loc[:, ['QoE', 'delay', 'jitter', 'loss']] 

data_merged.head(5)
```
